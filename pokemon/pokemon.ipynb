```python
for group , grouped in  pokedex.groupby('Type 1')['HP', 'Attack', 'Defense', 'Speed', 'Sp. Atk', 'Sp. Def']:
    grouped.drop(['#', 'Generation', 'Legendary', 'Total'], axis = 1).mean().plot.line(figsize = (5, 5), title = group)
    min_ = grouped.loc[:, 'HP':'Speed'].mean().min()
    plt.gca().set_ylim(bottom=min_)
    plt.show()
```
